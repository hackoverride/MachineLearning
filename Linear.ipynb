```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import mglearn
from IPython.display import display
```

```python
mglearn.plots.plot_linear_regression_wave()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

X, y = mglearn.datasets.make_wave(n_samples=60)
X_train,X_test, y_train, y_test = train_test_split(X, y, random_state=42)

lr = LinearRegression().fit(X_train, y_train)
```

```python
print("lr.coef_:", lr.coef_)
print("lr.intercept_:", lr.intercept_)
```

```python
#check test and training scores
print("Training score: {:.3f}".format(lr.score(X_train, y_train)))
print("Test score: {:.3f}".format(lr.score(X_test, y_test)))
#I like 3 point decimals better than two. dont know why ;)
```

```python
#When the test and trainingscore is very close together we are most likely underfitting...
```

```python
X, y = mglearn.datasets.load_extended_boston()
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
lr = LinearRegression().fit(X_train, y_train)

print("Test score: {:.3f}".format(lr.score(X_test, y_test)))
print("Training score: {:.3f}".format(lr.score(X_train, y_train)))
```

```python
from sklearn.linear_model import Ridge
ridge = Ridge().fit(X_train, y_train)

print("training score: {:.3f}".format(ridge.score(X_train, y_train)))
print("test score: {:.3f}".format(ridge.score(X_test, y_test)))
```

```python
mglearn.plots.plot_ridge_n_samples()
```

```python
from sklearn.linear_model import Lasso

lasso = Lasso().fit(X_train, y_train)
print("Training score: {:.3f}".format(lasso.score(X_train, y_train)))
print("Test score: {:.3f}".format(lasso.score(X_test, y_test)))
print("Number of features used:", np.sum(lasso.coef_ != 0))
```

```python
#default in Ridge and Lasso is alpha = 1
#To improve underfitting (low score) we can try to decrease the alpha

lasso = Lasso(alpha=0.05).fit(X_train, y_train)
print("Training score: {:.3f}".format(lasso.score(X_train, y_train)))
print("Test score: {:.3f}".format(lasso.score(X_test, y_test)))
print("Number of features used:", np.sum(lasso.coef_ != 0))
```

```python
#Going too low on alpha we need to make sure the iteration does not go out of bounds and cause the model to overfit
#Overfit = training is much higher than test because it is not doint what its ment to do.

lasso = Lasso(alpha=0.0001, max_iter=100000).fit(X_train, y_train)
print("Training score: {:.3f}".format(lasso.score(X_train, y_train)))
print("Test score: {:.3f}".format(lasso.score(X_test, y_test)))
print("Number of features used:", np.sum(lasso.coef_ != 0))
```

```python
#LogiticRegression
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, stratify=cancer.target, random_state=42)
logreg = LogisticRegression(max_iter=100000).fit(X_train, y_train)
logreg100 = LogisticRegression(C=100, max_iter=100000).fit(X_train, y_train)
logreg001 = LogisticRegression(C=0.01, max_iter=100000).fit(X_train, y_train)
print("training score: {:.3f}".format(logreg.score(X_train, y_train)))
print("training score: {:.3f}".format(logreg100.score(X_train, y_train)))
print("training score: {:.3f}".format(logreg001.score(X_train, y_train)))
```

```python
plt.plot(logreg.coef_.T, 'o', label="C=1")
plt.plot(logreg100.coef_.T, 'o', label="C=100")
plt.plot(logreg001.coef_.T, 'o', label="C=0.01")
plt.xticks(range(cancer.data.shape[1]), cancer.feature_names, rotation=90)
plt.hlines(0, 0, cancer.data.shape[1])
plt.ylim(-5, 5)
plt.xlabel("Feature")
plt.ylabel("Coefficient magnitude")
plt.legend()
```
