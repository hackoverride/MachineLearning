```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
cancer = load_breast_cancer()

X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, random_state=1)

print(X_train.shape)
print(X_test.shape)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaler.fit(X_train)
```

```python
X_train_scaled = scaler.transform(X_train)
print ("transformed shape : {}".format(X_train_scaled.shape))

#before
print("per-feature minimum before scaling:\n {}".format(X_train.min(axis=0)))
print("per-feature maximum before scaling:\n {}".format(X_train.max(axis=0)))
#after
print("per-feature minimum after scaling:\n {}".format(X_train_scaled.min(axis=0)))
print("per-feature maximum after scaling:\n {}".format(X_train_scaled.max(axis=0)))
```

```python
#Transforming the test data
X_test_scaled = scaler.transform(X_test)

print("per-f min after:\n {}".format(X_test_scaled.min(axis=0)))
print("per-f max after:\n {}".format(X_test_scaled.max(axis=0)))
```

```python
#a lot of code to show the visual data
from sklearn.datasets import make_blobs
X, _ = make_blobs(n_samples=50, centers=5, random_state=4, cluster_std=2)
X_train, X_test = train_test_split(X, random_state=5, test_size=.1)

#the fig, axes must be in the same "input range" as the axes[pos]
fig, axes= plt.subplots(1, 3, figsize=(13,4))


axes[0].scatter(X_train[:,0], X_train[:,1],
               color="red", label="Training set", s=60)

axes[0].scatter(X_test[:,0], X_test[:,1], marker='^',
               color="blue", label="Test set", s=60)
axes[0].legend(loc='upper left')
axes[0].set_title("Original Data")

#Scale the data
scaler = MinMaxScaler()
scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)

axes[1].scatter(X_train_scaled[:, 0], X_train_scaled[:, 1], color="red", label="Training", s=60)
axes[1].scatter(X_test_scaled[:, 0], X_test_scaled[:, 1], marker='^', color="blue", label="Test", s=60)
axes[1].set_title("Scaled Data");

#Rescale the test set separately

#this part is actually in the book warning that it is not supposed to do because we change the test data!
test_scaler = MinMaxScaler()
test_scaler.fit(X_test)
X_test_scaled_badly = test_scaler.transform(X_test)

#visualize in the right-most figure the bad data
axes[2].scatter(X_train_scaled[:, 0], X_train_scaled[:, 1], color="red", label="Training", s=60)
axes[2].scatter(X_test_scaled_badly[:, 1], X_test_scaled_badly[:, 1],marker='^', color="blue", label="test", s=60)
axes[2].legend(loc='upper left')
axes[2].set_title("improperly scaled data")

for ax in axes:
        ax.set_xlabel("Feature 0")
        ax.set_ylabel("Feature 1")
        
fig.tight_layout()
```

```python
# New heading : The effect of preprocessing on supervised learning

from sklearn.svm import SVC

X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, random_state=0)

svm = SVC(C=100)
svm.fit(X_train, y_train)
print("test accuracy: {:.2f}".format(svm.score(X_test, y_test)))
```

```python
#PCA finds the component direction (component1) and find that vector to normalize (transformed data)

#Apply PCA to the canacer dataset for visualization

fig, axes = plt.subplots(15, 2, figsize=(10, 20))
malignant = cancer.data[cancer.target == 0]
benign = cancer.data[cancer.target == 1]

ax = axes.ravel()

for i in range(30):
    _, bins = np.histogram(cancer.data[:, i], bins=50)
    ax[i].hist(malignant[:, i], bins=bins, color="red", alpha=.5)
    ax[i].hist(benign[:, i], bins=bins, color="blue", alpha=.5)
    ax[i].set_title(cancer.feature_names[i])
    ax[i].set_yticks(())

ax[0].set_xlabel("Feature magnitude")
ax[0].set_ylabel("Frequency")
ax[0].legend(["Malignant", "Bening"], loc="best")
fig.tight_layout()
```

```python
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
cancer = load_breast_cancer()

scaler = StandardScaler()
scaler.fit(cancer.data)
X_scaled = scaler.transform(cancer.data)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca.fit(X_scaled)
          
X_pca = pca.transform(X_scaled)
print("original: {}".format(str(X_scaled.shape)))
print("reduced: {}".format(str(X_pca.shape)))
```

```python
plt.figure(figsize=(8, 8))
mglearn.discrete_scatter(X_pca[:, 0], X_pca[:, 1], cancer.target)

plt.legend(cancer.target_names, loc="best")
plt.gca().set_aspect("equal")
plt.xlabel("first principal component")
plt.ylabel("Second principal componenet")
```
